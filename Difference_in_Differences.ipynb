```python
df = sm.add_constant(df)
```

```python
#DID year 1981 sample

X_1   = df[['const', 'y81', 'nearinc', 'y81nrinc']]
Y_1   = df[['lrprice']]
M1 = sm.OLS(Y_1, X_1, missing='drop')
OLS1 = M1.fit()
print(OLS1.summary())
```

```python
#DID
#improve the estimation by replacing nearinc by the variable dist, 
#Which is the distance from a house to the incinerator site
#log(price) = β0 + δ0y81 + β1log(dist) + δ1 ·y81 ·log(dist) + ε.


X_2 = df[['const', 'y81', 'ldist', 'y81ldist']]
Y_2 = df[['lprice']]
M2  = sm.OLS(Y_2, X_2, missing='drop')

OLS2 = M2.fit()
print(OLS2.summary())
```

```python

#Add age, age2, rooms, baths, log(intst), log(land), and log(area) to 4,
#where intst is the distance to interstate in feet


X_3 = df[['const', 'y81', 'ldist', 'y81ldist', 'age', 'agesq',
          'rooms', 'baths', 'lintst', 'lland', 'larea']]
Y_3 = df[['lprice']]
M3  = sm.OLS(Y_3, X_3, missing='drop')

OLS3 = M3.fit()
print(OLS3.summary())
```
